```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
# import files
df = pd.read_csv("C:/Users/kfeng3.DPU/Desktop/movie.csv")
df.head()
df.info()
```

```python
#Check Missing Value
df.isnull().sum()
df.duplicated().sum()
```

```python
df.dropna()
```

```python
#Check Zero Value
df["runtime"]=df["runtime"].replace(0,np.NaN)
df["vote_average"]=df["vote_average"].replace(0,np.NaN)
df["vote_count"]=df["vote_count"].replace(0,np.NaN)
df["popularity"]=df["popularity"].replace(0,np.NaN)
```

```python
df.isnull().sum()
```

```python
df1=df.dropna()
```

```python
df1.info()
```

```python
#Boxplot
sns.boxplot(df1["budget"])
```

```python
sns.boxplot(df1["revenue"])
```

```python
df1["budget"]=df1["budget"].replace(0, np.NaN)
df1["revenue"]=df1["revenue"].replace(0, np.NaN)
```

```python
df1["budget"].fillna(df1["budget"].median(), inplace=True)
df1["revenue"].fillna(df1["revenue"].median(), inplace=True)
```

```python
df1.to_csv("cleanfile.csv", encoding="utf-8")
```
